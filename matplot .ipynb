```python
x=np.array([0,1,2,3,])
y=np.array([0,8,2,6,])
mylabel=["apple","mango","orange","banaba"]
plt.pie(y,labels=mylabel)
plt.show()
```

```python
x=np.array([0,1,2,3,])
y=np.array([0,8,2,6,])
mylabel=["apple","mango","orange","banaba"]
plt.pie(y,labels=mylabel,startangle=90)
plt.show()
```

```python

y=np.array([1,8,2,6,])
mylabel=["apple","mango","orange","banaba"]
myexplode=[0.2,0,0,0]
plt.pie(y,labels=mylabel,startangle=90,explode=myexplode)#explode ka mains hai ki part ko alag karna 
plt.show()
```

```python
y=np.array([1,8,2,6,])
mylabel=["apple","mango","orange","banaba"]
myexplode=[0.2,0,0,0.9]
plt.pie(y,labels=mylabel,startangle=90,explode=myexplode,shadow=True)
plt.show()
```

```python
y=np.array([1,8,2,6,])
mylabel=["co2","n2","o2","caco3"]
myexplode=[0.2,0,0,0]
mycolor=['red','orange','pink','black']
plt.pie(y,labels=mylabel,startangle=90,explode=myexplode,shadow=True,colors=mycolor,autopct='%1.1f%%')
plt.show()
```

```python
y=np.array([1,8,2,6,])
mylabel=["co2","n2","o2","caco3"]
myexplode=[0.2,0,0,0]
mycolor=['red','orange','pink','black']
plt.pie(y,labels=mylabel,startangle=90,explode=myexplode,shadow=True,colors=mycolor,autopct='%1.1f%%')
plt.legend(title='color')
plt.show()
```
