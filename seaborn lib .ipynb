```python
sns.relplot(x='total_bill',y='tip',hue='time',data=data)#hue colors
```

```python

sns.relplot(x='total_bill',y='tip',hue='time',style='smoker',data=data,kind='scatter',col='smoker')
# kind is used two graph ,col is used ratio
        
```

```python
sns.relplot(x='total_bill',y='tip',hue='time',data=data,kind='scatter',style='smoker',row='time')
```
